# Collect with soup

This jupyter notebook is looping through the data folder

## 1. Collect data

```python
import glob
from bs4 import BeautifulSoup

# load html files
files = glob.glob('/Users/quinten/Documents/Universiteit Utrecht/Personalisation for (public) media (INFOMPPM)/Individual Assignment/data/NPO/shows/*.html', recursive=True)

data = []

for file in files:
  if './data/show' not in file:
    soup = BeautifulSoup(open(file), 'html.parser')
    article = {}
    print(file)
    
    # extract the data from each file
    if soup.find('meta',  attrs={'property': 'og:title'}):
      title = soup.find('meta',  attrs={'property': 'og:title'})['content']
      title = title.replace('gemist? Start met kijken op NPO Start', '')
    else:
      title = ''
    
    if soup.find('meta',  attrs={'property': 'og:description'}):
      description = soup.find('meta',  attrs={'property': 'og:description'})['content']
    else:
      description = ''

    if soup.find('meta',  attrs={'property': 'og:image'}):
      image = soup.find('meta',  attrs={'property': 'og:image'})['content']
    else:
      image = ''

    if soup.find('meta',  attrs={'property': 'al:ios:url'}):
      url = soup.find('meta',  attrs={'property': 'al:ios:url'})['content']
    else:
      url = ''

    article = {
      'title': title,
      'description': description,
      'image': image,
      'url': url
    }
    
    # append the article to the data
    data.append(article)
```

## 2. Saving the data

```python
import json
with open('data.json', 'w') as f:
    json.dump(data, f)
```
